```python
from targets.conditional_density_estimation_target import *
from models import CNDRE
from posterior_sampler import WLBSampler
```

```python
target = Wave()
target.target_visual()
num_samples = 50000
theta, x = target.make_dataset(num_samples)
```

```python
cndre = CNDRE(x, theta, [50,50,50], mode = 'Proxy')
cndre.train(epochs = 200, batch_size = 10000)
```

```python
import matplotlib.pyplot as plt
#x0
if x.shape[-1]==1 and theta.shape[-1]==1:
    theta_0 = torch.tensor([-2.])
    num_samples = 100
    x0 = target.simulate(theta_0.unsqueeze(0).repeat(num_samples,1))
    plt.hist(x0.numpy(),bins = 100, color = 'blue')
```

```python
%load_ext autoreload
%autoreload 2
wlb = WLBSampler(log_likelihood_function = cndre.log_density,x0 = x0, prior_sampler = target.sample_prior)
samples = wlb.sample_new(2000)
```

```python
linspace = 2000
theta_linspace = torch.linspace(-9., 9., linspace).unsqueeze(-1)
log_posterior = target.prior_log_prob(theta_linspace) + torch.sum(cndre.log_density(x0.unsqueeze(1).repeat(1,linspace,1),theta_linspace.unsqueeze(0).repeat(num_samples,1,1)), dim= 0)
hist = plt.hist(samples.detach().numpy(), bins = 75, density = True)
plt.figure()
plt.plot(theta_linspace.squeeze(-1),torch.exp(log_posterior).detach().numpy())
```

```python
theta_linspace = torch.linspace(-8.,8.,200)
x_linspace = torch.linspace(-2.,2.,200)
grid_theta,grid_x = torch.meshgrid(theta_linspace, x_linspace, indexing = 'ij')
density = cndre.log_density(grid_x.unsqueeze(-1), grid_theta.unsqueeze(-1))
plt.pcolormesh(grid_theta, grid_x,torch.exp(density).detach().numpy())
```
